```python
import matplotlib.pyplot as plt

# Data
references = ['[5]', '[6]', '[9]', '[11]', 'Proposed']
accuracy = [75, 70, 78, 83, 90]

# Colors (last one green for proposed)
colors = ['blue', 'blue', 'blue', 'blue', 'green']

# Plot
plt.figure(figsize=(8,5))
bars = plt.bar(references, accuracy, color=colors)

# Labels
plt.title("Prediction accuracy by reference")
plt.xlabel("Reference")
plt.ylabel("Accuracy (%)")

# Add values on top of bars
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 1,
             f'{height}%', ha='center')

# Y-axis limit
plt.ylim(60, 100)
plt.savefig("histogram.png", dpi=300, bbox_inches='tight')
plt.show()
```
